```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"
```

```python
#Read City data
city_data_df=pd.read_csv(city_data_to_load)
city_data_df.head(10)
```

```python
#read ride data
ride_data_df = pd.read_csv(ride_data_to_load)
ride_data_df.head(10)
```

```python
#City data no nulls
print(city_data_df.count())
#Shows no 0
#Data types
city_data_df.dtypes
```

```python
#Unique values
city_data_df["type"].unique()
```

```python
#How many are Urban/Sub/Rural?
print(sum(city_data_df["type"]=="Urban"))
print(sum(city_data_df["type"]=="Suburban"))
print(sum(city_data_df["type"]=="Rural"))
```

```python
#Look at ride data
print(ride_data_df.count())
ride_data_df.isnull().sum()
#Confirmation no null
```

```python
#Types
ride_data_df.dtypes
```

```python
#Merge data
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

pyber_data_df.head()
```

# STARTING 5.3.1

```python
#creating dataframes for all types
urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]
suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]
rural_cities_df.head()
```

```python
#number of rides for urban
urban_ride_count=urban_cities_df.groupby(['city']).count()['ride_id']
urban_ride_count.head()
```

```python
#number of rides for Sub/Rural
suburban_ride_count=suburban_cities_df.groupby(['city']).count()['ride_id']
rural_ride_count=rural_cities_df.groupby(['city']).count()['ride_id']
```

```python
# Mean of fare in Urban
urban_avg_fare = urban_cities_df.groupby(["city"]).mean()["fare"]
urban_avg_fare.head()
```

```python
#mean of fare in Sub/rural
suburban_avg_fare = suburban_cities_df.groupby(["city"]).mean()["fare"]
rural_avg_fare = rural_cities_df.groupby(["city"]).mean()["fare"]
```

```python
#mean number of drivers Urban 
urban_driver_count = urban_cities_df.groupby(["city"]).mean()["driver_count"]
urban_driver_count
```

```python
#mean number of drivers sub/rural 
suburban_driver_count = suburban_cities_df.groupby(["city"]).mean()["driver_count"]
rural_driver_count = rural_cities_df.groupby(["city"]).mean()["driver_count"]
```

```python
#Make a scatter plot for urban
plt.scatter(urban_ride_count, 
            urban_avg_fare,
           s=10*urban_driver_count,
           c="coral",
           edgecolor="black",
           linewidths=1,
           alpha=0.8,#alpha makes things transparent
           label="Urban")
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
plt.legend()
```

```python
#Make a scatter plot for suburban
plt.scatter(suburban_ride_count, 
            suburban_avg_fare,
           s=10*suburban_driver_count,
           c="skyblue",
           edgecolor="black",
           linewidths=1,
           alpha=0.8,#alpha makes things transparent
           label="Suburban")
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
plt.legend()
```

```python
#Make a scatter plot for Rural
plt.scatter(rural_ride_count, 
            rural_avg_fare,
           s=10*rural_driver_count,
           c="gold",
           edgecolor="black",
           linewidths=1,
           alpha=0.8,#alpha makes things transparent
           label="Rural")
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
plt.legend()
```

```python
#Combining all plots
plt.scatter(rural_ride_count, 
            rural_avg_fare,
           s=10*rural_driver_count,
           c="gold",
           edgecolor="black",
           linewidths=1,
           alpha=0.8,#alpha makes things transparent
           label="Rural")
plt.scatter(suburban_ride_count, 
            suburban_avg_fare,
           s=10*suburban_driver_count,
           c="skyblue",
           edgecolor="black",
           linewidths=1,
           alpha=0.8,#alpha makes things transparent
           label="Suburban")
plt.scatter(urban_ride_count, 
            urban_avg_fare,
           s=10*urban_driver_count,
           c="coral",
           edgecolor="black",
           linewidths=1,
           alpha=0.8,#alpha makes things transparent
           label="Urban")

plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
lndg=plt.legend(fontsize="12", mode="Expanded",
         scatterpoints=1, loc="best", title="City Types")
lgnd.legendHandles[0]._sizes = [75]
lgnd.legendHandles[1]._sizes = [75]
lgnd.legendHandles[2]._sizes = [75]
lgnd.get_title().set_fontsize(12)
#All above needs to go ABOVE show
plt.show()

```
